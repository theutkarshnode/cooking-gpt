### Implement Chatbot Logic

The following code implements the `send_message_to_chatbot` function. This function takes a user's query, initializes a chat session with the generative model, provides system instructions to guide the model's persona, sends the user's message, and then returns the AI-generated response. This will be the core of our conversational interface for the cooking chatbot.

**Reasoning**:
I need to implement the `send_message_to_chatbot` function as described in the instructions to fulfill the subtask of developing the chatbot's core logic. This function will initiate a chat, set the persona, send the user query, and return the model's response.



```python
def send_message_to_chatbot(user_query):
    chat = model.start_chat(history=[], enable_automatic_function_calling=True)
    chat.send_message(f"system_instruction: Act as a helpful cooking assistant.")
    response = chat.send_message(user_query)
    return response.text

print("Chatbot function 'send_message_to_chatbot' defined.")
```
